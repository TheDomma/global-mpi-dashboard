```python
import pandas as pd

# 1. Load the raw dataset
print("Loading raw data...")
df = pd.read_csv('global_mpi.csv')

# Show original shape and missing values
print(f"Original Data Shape: {df.shape}")
print("\nMissing values per column before cleaning:")
print(df.isnull().sum())

# 2. Data Cleaning Steps
# Drop rows where critical metrics (MPI or Intensity of Deprivation) are missing
df_cleaned = df.dropna(subset=['MPI', 'Intensity of Deprivation']).copy()

# Fill missing Admin 1 Names (Regions) with 'National Average' or 'Unknown'
# The dataset has some rows that represent the whole country rather than a specific region
df_cleaned['Admin 1 Name'] = df_cleaned['Admin 1 Name'].fillna('National Average')

# Convert Start and End Dates to proper datetime objects
df_cleaned['Start Date'] = pd.to_datetime(df_cleaned['Start Date'])
df_cleaned['End Date'] = pd.to_datetime(df_cleaned['End Date'])

# 3. Verify Cleaning
print(f"\nCleaned Data Shape: {df_cleaned.shape}")
print("\nMissing values per column after cleaning:")
print(df_cleaned.isnull().sum())

# 4. Export the Cleaned Dataset
df_cleaned.to_csv('cleaned_global_mpi.csv', index=False)
print("\nSuccess! Saved as 'cleaned_global_mpi.csv'")
```
