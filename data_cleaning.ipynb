```python
import pandas as pd
import pycountry

# 1. Load the raw dataset
print("Loading raw data...")
df = pd.read_csv('global_mpi.csv')

# 2. Data Cleaning Steps
# Step 2a: Remove any exact duplicate rows
df_cleaned = df.drop_duplicates().copy()

# Step 2b: Drop rows where critical metrics are missing
df_cleaned = df_cleaned.dropna(subset=['MPI', 'Intensity of Deprivation'])

# Step 2c: Drop the 'Admin 1 PCode' column as it has too many nulls and is unused
df_cleaned = df_cleaned.drop(columns=['Admin 1 PCode'])

# Step 2d: Fill missing Admin 1 Names with 'National Average'
df_cleaned['Admin 1 Name'] = df_cleaned['Admin 1 Name'].fillna('National Average')

# Step 2e: Convert dates to datetime objects
df_cleaned['Start Date'] = pd.to_datetime(df_cleaned['Start Date'])
df_cleaned['End Date'] = pd.to_datetime(df_cleaned['End Date'])

# 3. Feature Engineering: Convert ISO3 Codes to Full Country Names
print("Converting ISO3 codes to Country Names...")
def get_country_name(iso_code):
    try:
        return pycountry.countries.get(alpha_3=iso_code).name
    except:
        return iso_code # Fallback

# Create the new 'Country' column
df_cleaned['Country'] = df_cleaned['Country ISO3'].apply(get_country_name)

# Reorder columns so 'Country' is at the front next to 'Country ISO3'
cols = df_cleaned.columns.tolist()
cols = cols[-1:] + cols[:-1]
df_cleaned = df_cleaned[cols]

# 4. Export the Cleaned Dataset
df_cleaned.to_csv('cleaned_global_mpi.csv', index=False)
print("\nSuccess! Saved as 'cleaned_global_mpi.csv'. The dataset is fully cleaned and ready for the dashboard!")
```
